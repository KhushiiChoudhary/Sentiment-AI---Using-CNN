```python
import numpy as np
import pandas as pd
from nltk.corpus import stopwords
import re
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
DATASET_COLUMNS = ["target", "ids", "date", "flag", "user", "text"]
DATASET_ENCODING = "ISO-8859-1"
data = pd.read_csv('training.1600000.processed.noemoticon.csv', encoding =DATASET_ENCODING , names=DATASET_COLUMNS)
data.head()
X = data.iloc[:,[5]]
Y = data.iloc[:,0]
Y = Y.copy()
Y.loc[Y == 4] = 1
```

```python
data.head()
```

```python
import nltk

nltk.download('stopwords')
```

```python
# Text-preprocessing

# Missing Values
num_missing_desc = data.isnull().sum()[2]  # No. of values with missing descriptions
print('Number of missing values: ' + str(num_missing_desc))
data = data.dropna()

TAG_CLEANING_RE = "@\S+"
# Remove @tags
X['text'] = X['text'].apply(lambda x: re.sub(TAG_CLEANING_RE, ' ', x) if isinstance(x, str) else x)

# Smart lowercase
X['text'] = X['text'].apply(lambda x: x.lower() if isinstance(x, str) else x)

# Remove numbers
X['text'] = X['text'].apply(lambda x: re.sub(r'\d+', ' ', x) if isinstance(x, str) else x)

# Remove links
TEXT_CLEANING_RE = "https?:\S+|http?:\S|[^A-Za-z0-9]+"
X['text'] = X['text'].apply(lambda x: re.sub(TEXT_CLEANING_RE, ' ', x) if isinstance(x, str) else x)

# Remove Punctuation
X['text'] = X['text'].apply(lambda x: x.translate(x.maketrans('', '', string.punctuation)) if isinstance(x, str) else x)

# Remove white spaces
X['text'] = X['text'].apply(lambda x: x.strip() if isinstance(x, str) else x)

# Tokenize into words
X['text'] = X['text'].apply(lambda x: word_tokenize(x) if isinstance(x, str) else x)

# Remove non-alphabetic tokens
X['text'] = X['text'].apply(lambda x: [word for word in x if word.isalpha()] if isinstance(x, list) else x)

# Filter out stop words
stop_words = set(stopwords.words('english'))
X['text'] = X['text'].apply(lambda x: [w for w in x if not w in stop_words] if isinstance(x, list) else x)

# Word Lemmatization
lem = WordNetLemmatizer()
X['text'] = X['text'].apply(lambda x: [lem.lemmatize(word, "v") for word in x] if isinstance(x, list) else x)

# Turn lists back to string
X['text'] = X['text'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
print("TRAIN size:", len(X_train))
print("TEST size:", len(X_train))
```

```python
# Word2Vec
import gensim

# WORD2VEC 
W2V_VECTOR_SIZE = 300
W2V_WINDOW = 7
W2V_EPOCH = 32
W2V_MIN_COUNT = 10

# Assuming X_train.text is already a list of strings
documents = X_train.text

w2v_model = gensim.models.Word2Vec(vector_size=W2V_VECTOR_SIZE, 
                                   window=W2V_WINDOW, 
                                   min_count=W2V_MIN_COUNT, 
                                   workers=8)
w2v_model.build_vocab(documents)
```

```python
words = list(w2v_model.wv.index_to_key)
vocab_size = len(words)
print("Vocab size:", vocab_size)
```

```python
# Train Word Embeddings
w2v_model.train(documents, total_examples=len(documents), epochs=W2V_EPOCH)
```
